## 1. Connect to Arvest

```python
ar = arvestapi.Arvest("jacob.hart@univ-rennes2.fr", "arvestfac1")

print(ar.profile.name)
```

## 2. Get my content

```python
medias = ar.get_medias()

for item in medias:
  print(item.title)
  print(item.description)
  print(item.url)

print()

manifests = ar.get_manifests()
for item in manifests:
  print(item.title)
print()

projects = ar.get_projects()
for item in projects:
  print(item.title)
```

## 3. Add new content

```python
import os

local_path = os.path.join(os.getcwd(), "moleman copie.jpeg")
image_url = "https://i.pinimg.com/originals/fd/3b/78/fd3b78ab6e60a8a9b3d45073c4b8fb95.jpg"

# Add local file:
new_media = ar.add_media(path = local_path)
new_media.update_title("NOUVEAU TITRE")
new_media.update_description("")

# Add online file:
ar.add_media(path = image_url)
```

```python
ar.create_project(title = "MON NOUVEAU PROJET")
```

## 4. Metadata

```python
medias = ar.get_medias()

for item in medias:
  if item.title == "Dice":
    metadata = item.get_metadata()
    print(metadata)
```

```python
import iiif_prezi3
import os
import json

m = iiif_prezi3.Manifest(id="https://example.com/iiif/manifest.json", label={"en":["Example Manifest"]})

print(m.dict())

out = os.path.join(os.getcwd(), "output.json")

with open(out, 'w', encoding='utf-8') as f:
  json.dump(m.dict(), f, ensure_ascii = False, indent = 2)
```

```python
c = iiif_prezi3.Canvas()

m.items.append(c)

with open(out, 'w', encoding='utf-8') as f:
  json.dump(m.dict(), f, ensure_ascii = False, indent = 2)
```
