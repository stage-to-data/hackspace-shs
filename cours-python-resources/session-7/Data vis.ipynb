```python
import os

chemin = os.path.join(os.getcwd(), "basic-data.csv")

print(chemin)
```

```python
import csv

donnes = []

with open(chemin) as f:
  csv_reader = csv.reader(f)
  for row in csv_reader:
    donnes.append(row)

print(donnes)
```

```python
import matplotlib.pyplot as plt

plt.plot(donnes[0], donnes[1])
plt.title("My amazing plot")
plt.xlabel("Bonjour")
plt.ylabel("Hello")

plt.show()
```

```python
plt.scatter(donnes[0], donnes[1])
plt.show()
```

```python
plt.pie(donnes[1], labels = donnes[0])
plt.savefig("OUT.png")
plt.show()
```

```python
plt.bar(donnes[0], donnes[1])
plt.show()
```

```python
graph, subplot = plt.subplots(1, 1)

subplot.plot(donnes[0], donnes[1])
subplot.set_title("Mon subplot")
subplot.invert_yaxis()

graph.tight_layout()

plt.show()
```

```python
import pandas as pd

chemin = os.path.join(os.getcwd(), "other_data.csv")

data = pd.read_csv(chemin, header = None, names = ['x', 'y', 'class'])

print(data)
```

```python
color_map = {
    1 : "green",
    2 : 'blue',
    3 : "red"
}

plt.figure(figsize=(8, 8))

for class_label, group in data.groupby("class"):
  plt.scatter(
      group["x"],
      group["y"],
      label=f"Class {class_label}",
      color = color_map.get(class_label, "black")
    )

plt.legend()
plt.show()
```
