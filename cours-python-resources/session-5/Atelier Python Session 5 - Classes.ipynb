## Creation de classe

```python
# Pour creer une classe, on utilise le mot clé class, puis le nom (avec un majuscule)
class Etudiant:

  # On doit créer une methode __init__() avec au mopins l'argument self.
  def __init__(self, prenom_etudiant : str, nom_de_famille : str, math : float, francais : float):
    """Represente un etudiant lambda """
    self.prenom = prenom_etudiant
    self.nom_de_famille = nom_de_famille
    self.note_maths = math
    self.note_francais = francais

  def get_moyenne(self):
    moyenne = (self.note_maths + self.note_francais) / 2
    return moyenne

class EtudiantSpeAnglais(Etudiant):

  def __init__(self, prenom_etudiant : str, nom_de_famille : str, math : float, francais : float, anglais : float):
    super().__init__(prenom_etudiant, nom_de_famille, math, francais)
    self.note_anglais = anglais

  def get_moyenne(self):
    moy_super = super().get_moyenne()
    return ((moy_super * 2) + self.note_anglais) / 3
```

```python
etudiant_3 = EtudiantSpeAnglais("Jeanne", "Fras", 9, 14, 18)

print(etudiant_3.nom_de_famille)

print(etudiant_3.get_moyenne())
```

## Creer une instance de la classe

```python
etudiant_1 = Etudiant("Jacob", "Hart", 12, 10)
etudiant_2 = Etudiant("Clarisse", "Bardiot", 16, 20)

print(etudiant_1.note_francais)
print(etudiant_2.note_francais)
```

On peut acceder à des attributs de la classe de cette manière:

```python
print(etudiant_1.prenom)
```

## Les methodes

```python
moyenne = (etudiant_1.note_francais + etudiant_1.note_maths) / 2
moyenne2 = (etudiant_2.note_francais + etudiant_2.note_maths) / 2

print(moyenne)
print(moyenne2)
```

```python
moyenne = etudiant_1.get_moyenne()
moyenne2 = etudiant_2.get_moyenne()

print(moyenne)
print(moyenne2)

print(etudiant_1.get_moyenne())
```
